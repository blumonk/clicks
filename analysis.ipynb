```python
referers = clicks.groupby('referer')['book'].mean().sort_values(ascending=False)
referers.plot(kind='bar', title='Referers: orders per click');
```

```python
most_popular = clicks[clicks.book == 1]\
                .groupby(['iso_from', 'iso_to'])\
                .size()\
                .sort_values()[-50:]\
                .reset_index(name='books')
most_popular['from_to'] = most_popular.iso_from.map(str) + '-' + most_popular.iso_to
top20 = most_popular.tail(20)[['books', 'from_to']]

fig = plt.figure(figsize=(10, 5))
plt.title('Top-20 most popular routes')
plt.barh(np.arange(len(top20)), top20.books)
plt.yticks(np.arange(len(top20)), top20.from_to)
plt.gca().set_xlabel('Total orders');
```

```python
top_opc = clicks.groupby(['iso_from', 'iso_to'])['book']\
                    .apply(lambda x: len(x[x == 1]) / len(x))
top_opc = top_opc.reset_index(name='opc')
top = pd.merge(most_popular, top_opc, on=['iso_from', 'iso_to']).iloc[::-1]
top.sort_values(by='opc', ascending=False, inplace=True)

fig = plt.figure(figsize=(12, 5))
plt.title('Top-50 most popular routes sorted by conversion')
plt.bar(np.arange(len(top)), top.opc)
plt.xticks(np.arange(len(top)), top.from_to, rotation=60);
plt.gca().set_ylabel('Orders per click');
```

```python
a_days = clicks[clicks.a_days >= 0].groupby('a_days')\
            .apply(lambda x: len(x[x.book == 1]) / len(x))
plt.figure(figsize=(10, 5))
ax = plt.gca()
a_days[:60].plot()
ax.set_ylabel('Orders per click')
ax.set_xlabel('Days till departure');
```

```python
a_days_books = clicks[clicks.book == 1].groupby('a_days').size()[:60]
plt.figure(figsize=(10, 5))
ax = plt.gca()

a_days_books.plot()
ax.set_ylabel('Total bookings')
ax.set_xlabel('Days till departure');
```

```python
import seaborn as sns
import matplotlib
from datetime import datetime as dt

#sns.distplot(clicks[clicks.book==1].date)
offset = 736000
dates = [dt.fromordinal(x + offset).strftime("%d.%m.%y") for x in range(0, 450, 50)]
plt.figure(figsize=(10, 5))

sns.distplot(clicks[clicks.book==1].date_from, color='red').set(xticklabels=dates)
g = sns.distplot(clicks[clicks.book==1].date_back, color='blue')
g.axes.set_xlim(offset,offset + 400)
g.set(xlabel='');

legend_patches = [matplotlib.patches.Patch(color=C, label=L) for
                  C, L in [('red', 'Departure'), ('blue', 'Arrival')]]
plt.legend(handles=legend_patches)
plt.title('Departure and arrival dates distribution');
```
