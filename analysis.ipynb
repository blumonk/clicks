# Basic analysis

```python
print('Clicks per purchase: %.3f' % (len(clicks)/len(clicks[clicks.book == 1])))
```

```python
referers = clicks.groupby('referer')['book'].mean().sort_values(ascending=False)
referers.plot(kind='bar', title='Referers: orders per click');
```

```python
most_popular = clicks[clicks.book == 1]\
                .groupby(['iso_from', 'iso_to'])\
                .size()\
                .sort_values()[-50:]\
                .reset_index(name='books')
most_popular['from_to'] = most_popular.iso_from.map(str) + '-' + most_popular.iso_to
top20 = most_popular.tail(20)[['books', 'from_to']]

fig = plt.figure(figsize=(10, 5))
plt.title('Top-20 most popular routes')
plt.barh(np.arange(len(top20)), top20.books)
plt.yticks(np.arange(len(top20)), top20.from_to)
plt.gca().set_xlabel('Total orders');
```

```python
top_opc = clicks.groupby(['iso_from', 'iso_to'])['book']\
                    .apply(lambda x: len(x[x == 1]) / len(x))
top_opc = top_opc.reset_index(name='opc')
top = pd.merge(most_popular, top_opc, on=['iso_from', 'iso_to']).iloc[::-1]
top.sort_values(by='opc', ascending=False, inplace=True)

fig = plt.figure(figsize=(12, 5))
plt.title('Top-50 most popular routes sorted by conversion')
plt.bar(np.arange(len(top)), top.opc)
plt.xticks(np.arange(len(top)), top.from_to, rotation=60);
plt.gca().set_ylabel('Orders per click');
```

```python
a_days = clicks[clicks.a_days >= 0].groupby('a_days')\
            .apply(lambda x: len(x[x.book == 1]) / len(x))
plt.figure(figsize=(10, 5))
ax = plt.gca()
a_days[:60].plot()
ax.set_ylabel('Orders per click')
ax.set_xlabel('Days till departure');
```

```python
a_days_books = clicks[clicks.book == 1].groupby('a_days').size()[:60]
plt.figure(figsize=(10, 5))
ax = plt.gca()

a_days_books.plot()
ax.set_ylabel('Total bookings')
ax.set_xlabel('Days till departure');
```

```python
import seaborn as sns
import matplotlib
from datetime import datetime as dt

offset = 736000
dates = [dt.fromordinal(x + offset).strftime("%d.%m.%y") for x in range(0, 450, 50)]
plt.figure(figsize=(10, 5))

sns.distplot(clicks[clicks.book==1].date_from, color='red').set(xticklabels=dates)
g = sns.distplot(clicks[clicks.book==1].date_back, color='blue')
g.axes.set_xlim(offset,offset + 400)
g.set(xlabel='');

legend_patches = [matplotlib.patches.Patch(color=C, label=L)
                  for C, L in [('red', 'Departure'), ('blue', 'Arrival')]]
plt.legend(handles=legend_patches)
plt.title('Departure and arrival dates distribution');
```

```python
categorical = ['iata_from', 'iata_to', 'iso_from', 'iso_to']
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16,10))
axes = [ax1, ax2, ax3, ax4]
cr = lambda x: len(x[x.book == 1]) / len(x)
for ax, feature in zip(axes, categorical):
    g = sns.distplot(clicks.groupby(feature).apply(cr), bins=np.arange(-0.2, 0.3, 0.005), ax=ax)
    g.set_xlim(-0.2, 0.4)
    ax.set_title('CR frequency grouped by {}'.format(feature))
    ax.set_xlabel('Conversion rate')
    ax.set_ylabel('Frequency')
```

```python
from matplotlib.patches import Patch

plt.figure(figsize=(10, 5))
columns = ['iata_from', 'iata_to', 'iso_from', 'iso_to']
colors = 'bgry'
alpha = 0.5

cr = lambda x: len(x[x.book == 1]) / len(x)
for column, color in zip(columns, colors):
    grouped = clicks.groupby(column)
    crs = grouped.apply(cr)
    ns = grouped.apply(len)

    cr_n = pd.concat([crs, ns], axis=1).reset_index().rename(columns={0: 'cr', 1: 'n'})
    cr_n.sort_values(by='cr', inplace=True)

    crs = np.arange(0, 0.7, 0.01)
    ns = []
    for x in crs:
        n = sum(cr_n[cr_n.cr < x].n)    
        ns.append(n)
    plt.plot(crs, ns, color=color, alpha=alpha)

text_patch = [Patch(color='none', label='Grouped by:')]
legend_patches = [Patch(color=C, label=L, alpha=alpha) for C, L in zip(colors, columns)]
plt.legend(handles=text_patch+legend_patches)
plt.title("Number of clicks with CR less than 'x'");
ax = plt.gca()
ax.set_ylabel('Number of clicks')
ax.set_xlabel('Conversion rate');
```
